```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
    query1 = """
        SELECT *
        FROM FACILITIES
        """
    cur.execute(query1)
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```

```python
#/* Q8: Produce a list of bookings on the day of 2012-09-14 which
#will cost the member (or guest) more than $30. Remember that guests have
#different costs to members (the listed costs are per half-hour 'slot'), and
#the guest user's ID is always 0. Include in your output the name of the
#facility, the name of the member formatted as a single column, and the cost.
#Order by descending cost, and do not use any subqueries. */


SELECT 
    f.name AS facility_name,
    CONCAT(m.firstname, ' ', m.surname) AS member_name,
    CASE 
        WHEN b.memid = 0 THEN (b.slots * f.guestcost)  -- Guests (ID = 0) cost
        ELSE (b.slots * f.membercost)                 -- Members cost
    END AS cost
FROM Bookings b
JOIN Facilities f ON b.facid = f.facid
JOIN Members m ON b.memid = m.memid
WHERE b.starttime LIKE '2012-09-14%'
  AND (CASE 
           WHEN b.memid = 0 THEN (b.slots * f.guestcost) 
           ELSE (b.slots * f.membercost) 
       END) > 30
ORDER BY cost DESC;


|Massage Room 2|GUEST GUEST|320.0
|Massage Room 1|GUEST GUEST|160.0
|Massage Room 1|GUEST GUEST|160.0
|Massage Room 1|GUEST GUEST|160.0
|Tennis Court 2|GUEST GUEST|150.0
|Tennis Court 1|GUEST GUEST|75.0
|Tennis Court 1|GUEST GUEST|75.0
|Tennis Court 2|GUEST GUEST|75.0
|Squash Court|GUEST GUEST|70.0
|Massage Room 1|Jemima Farrell|39.6
|Squash Court|GUEST GUEST|35.0
|Squash Court|GUEST GUEST|35.0

```

```python
/* Q9: This time, produce the same result as in Q8, but using a subquery. */

SELECT booking_details.facility_name,
       CONCAT(m.firstname, ' ', m.surname) AS member_name,
       booking_details.booking_cost AS cost
FROM (
         SELECT f.name AS facility_name,
                b.bookid,
                b.memid,
                CASE
                    WHEN b.memid = 0 THEN (b.slots * f.guestcost)  -- Cost for guests
                    ELSE (b.slots * f.membercost)                 -- Cost for members
                END AS booking_cost
         FROM Bookings b
         JOIN Facilities f ON b.facid = f.facid
         WHERE b.starttime LIKE '2012-09-14%'  -- Filter by date
     ) AS booking_details
JOIN Members m ON booking_details.memid = m.memid
WHERE booking_details.booking_cost > 30
ORDER BY booking_details.booking_cost DESC;


|Massage Room 2|GUEST GUEST|320.0
|Massage Room 1|GUEST GUEST|160.0
|Massage Room 1|GUEST GUEST|160.0
|Massage Room 1|GUEST GUEST|160.0
|Tennis Court 2|GUEST GUEST|150.0
|Tennis Court 1|GUEST GUEST|75.0
|Tennis Court 1|GUEST GUEST|75.0
|Tennis Court 2|GUEST GUEST|75.0
|Squash Court|GUEST GUEST|70.0
|Massage Room 1|Jemima Farrell|39.6
|Squash Court|GUEST GUEST|35.0
|Squash Court|GUEST GUEST|35.0


```

```python

#/* Q10: Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that there's a different cost for guests and members! */

import sqlite3
import os
import pandas as pd

# Absolute path to the database
db_path = os.path.abspath('sqlite_db_pythonsqlite.db')  # Adjust path to match your database location

try:
    # Connect to SQLite database
    conn = sqlite3.connect(db_path)
    print("Connection established successfully.")
    
    # Execute the SQL query (replace with your query as needed)
    query1 = '''
        SELECT f.name AS facility_name,
               SUM(CASE 
                       WHEN b.memid = 0 THEN b.slots * f.guestcost
                       ELSE b.slots * f.membercost
                   END) AS total_revenue
        FROM Bookings b
        JOIN Facilities f ON b.facid = f.facid
        GROUP BY f.name
        HAVING total_revenue < 1000
        ORDER BY total_revenue;
    '''
    
    # Fetch results into a pandas DataFrame
    result_df = pd.read_sql_query(query1, conn)
    print(result_df)
    
except sqlite3.Error as e:
    print(f"Database connection error: {e}")
finally:
    if conn:
        conn.close()
        print("Connection closed.")
```

```python
#/* Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order */
import sqlite3
import pandas as pd

try:
    # Connect to the SQLite database
    db_path = 'sqlite_db_pythonsqlite.db'  # Ensure this path matches your database location
    conn = sqlite3.connect(db_path)
    print("Connection established successfully.")
    
    # Define the SQL query
    query = '''
        SELECT m1.surname AS member_surname, 
               m1.firstname AS member_firstname,
               m2.surname AS recommender_surname, 
               m2.firstname AS recommender_firstname
        FROM Members m1
        LEFT JOIN Members m2 ON m1.recommendedby = m2.memid
        ORDER BY m1.surname, m1.firstname;
    '''
    
    # Execute the query and fetch the result into a pandas DataFrame
    result_df = pd.read_sql_query(query, conn)
    print(result_df)
    
except sqlite3.Error as e:
    print(f"Database connection error: {e}")
finally:
    if conn:
        conn.close()
        print("Connection closed.")

```

```python
#/* Q12: Find the facilities with their usage by member, but not guests */

import sqlite3
import pandas as pd

try:
    # Connect to the SQLite database
    db_path = 'sqlite_db_pythonsqlite.db'  # Ensure this path matches your database location
    conn = sqlite3.connect(db_path)
    print("Connection established successfully.")
    
    # Define the SQL query
    query = '''
        SELECT f.name AS facility_name,
               SUM(b.slots) AS total_usage_by_members
        FROM Bookings b
        JOIN Facilities f ON b.facid = f.facid
        WHERE b.memid != 0  -- Exclude guests (memid = 0)
        GROUP BY f.name
        ORDER BY total_usage_by_members DESC;
    '''
    
    # Execute the query and fetch the result into a pandas DataFrame
    result_df = pd.read_sql_query(query, conn)
    print(result_df)
    
except sqlite3.Error as e:
    print(f"Database connection error: {e}")
finally:
    if conn:
        conn.close()
        print("Connection closed.")
```

```python
#/* Q13: Find the facilities usage by month, but not guests */

import sqlite3
import pandas as pd

try:
    # Connect to the SQLite database
    db_path = 'sqlite_db_pythonsqlite.db'  # Ensure this path matches your database location
    conn = sqlite3.connect(db_path)
    print("Connection established successfully.")
    
    # Define the SQL query
    query = '''
        SELECT strftime('%Y-%m', b.starttime) AS month,  -- Extract year-month from booking start time
               f.name AS facility_name,
               SUM(b.slots) AS total_usage_by_members
        FROM Bookings b
        JOIN Facilities f ON b.facid = f.facid
        WHERE b.memid != 0  -- Exclude guests (memid = 0)
        GROUP BY month, f.name
        ORDER BY month, total_usage_by_members DESC;
    '''
    
    # Execute the query and fetch the result into a pandas DataFrame
    result_df = pd.read_sql_query(query, conn)
    print(result_df)
    
except sqlite3.Error as e:
    print(f"Database connection error: {e}")
finally:
    if conn:
        conn.close()
        print("Connection closed.")
```
